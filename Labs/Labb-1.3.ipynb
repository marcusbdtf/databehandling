## Exploring total number of cases in a 13 day xtick period in the 5 largest regions of sweden

```python
import plotly_express as px

fig = px.histogram(df, x = "Statistikdatum", y = ['Skåne', 'Stockholm',
             'Uppsala', 'Östergötland', 'Västra_Götaland'], 
             title = 'Antal fall 2020-02-04 - 2021-10-25 (5 största regionerna)',
             labels = {'Statistikdatum': 'Statistikdatum, x = 13 dagars intervall'})
fig.update_layout(legend_title = None,
                   hovermode = "x")                         
fig.update_xaxes(
    showspikes = True, 
    spikedash = "solid",
    spikemode = "across",
    spikecolor ="black",
    spikesnap = "cursor",
)       
fig.show()
```

- Cases started taking off in the Stockholm region and then Västra Götaland

```python
"""Exploring total cases in men/women and how many of them needed intensive care/died """

fig = px.bar(df1, x = 'Kön', y = ['Totalt_antal_fall', 'Totalt_antal_intensivvårdade',
       'Totalt_antal_avlidna'] )
fig.update_layout( barmode = 'group',
                   legend_title = None,
                   hovermode = "x")                         
fig.update_xaxes(
    showspikes = True, 
    spikedash = "solid",
    spikemode = "across",
    spikecolor ="black",
    spikesnap = "cursor",
)       
fig.show()
```

- We can see that even though women had more cases men still died at a higher rate with 1421 more deaths
- With 20k more cases women, men still had ~ 3k more cases of intensive care
